**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

**Importing the CSV file**

```python
df=pd.read_csv("/content/train.csv",index_col=0)
df
```

**Data Overview**

```python
df.info(verbose=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

**Data Cleaning**

```python
df.rename(columns = {'Order ID':'Order_ID'}, inplace = True)
df.rename(columns = {'Order Date':'Order_Date'}, inplace = True)
df.rename(columns = {'Ship Date':'Ship_Date'}, inplace = True)
df.rename(columns = {'Ship Mode':'Ship_Mode'}, inplace = True)
df.rename(columns = {'Customer ID':'Customer_ID'}, inplace = True)
df.rename(columns = {'Customer Name':'Customer_Name'}, inplace = True)
df.rename(columns = {'Postal Code':'Postal_Code'}, inplace = True)
df.rename(columns = {'Product ID':'Product_ID'}, inplace = True)
df.rename(columns = {'Sub-Category':'Sub_Category'}, inplace = True)
df.rename(columns = {'Product Name':'Product_Name'}, inplace = True)
```

```python
df['Order_Date']=pd.to_datetime(df.Order_Date )
```

**Visualization**

```python
#Histogram plot foe sales column
plt.figure(figsize=(10,6))
sns.histplot(df['Sales'],bins=10,kde=True)
plt.title("Count for Sales")
```

```python
#Count plot for Ship_Models column
plt.figure(figsize=(8,6))
sns.countplot(data=df,x="Ship_Mode",color="green")
plt.title("Count for Ship Models")
plt.xlabel("Ship Mode")
plt.ylabel("Count")
plt.show()
```

```python
#Pie chart for Region column
plt.figure(figsize=(15,6))
region=df['Region'].value_counts()
region.plot(kind='pie',autopct='%1.0f%%')
plt.title("Pie chart for Region")
```

```python
#Count plot for Category Column
plt.figure(figsize=(8,6))
sns.countplot(data=df,x="Category",color="c")
plt.title("Count for Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
#Bar plot for Year VS Sales Columns
plt.figure(figsize=(10,8))
df['Year']=df['Order_Date'].dt.year
cat=df[["Year", "Sales"]].groupby(['Year'], as_index=False).sum().sort_values(by='Sales', ascending=False)
sns.barplot(x='Year', y='Sales', data=cat,color="orange")
plt.grid()
plt.title("Sales by Years")
```

```python
#Pie chart for Category Column
plt.figure(figsize=(15,6))
region=df['Category'].value_counts()
region.plot(kind='pie',autopct='%1.0f%%')
plt.title("Pie chart for Category")
```

```python
#Bar plot for State VS Sales for top 20 Observation
top_state = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(20)
top_state = top_state.reset_index()
plt.figure(figsize=(15,6))
sns.barplot(data=df,x=top_state.Sales,y=top_state.State,color="red")
plt.title("Barplot for State VS Sales for top 20 Observations")

```

```python
#Barplot for Sales column monthly wise
plt.figure(figsize=(10,8))
df['Month']=df['Order_Date'].dt.month
cat=df[["Month", "Sales"]].groupby(['Month'], as_index=False).sum().sort_values(by='Sales', ascending=False)
sns.barplot(x='Month', y='Sales', data=cat,color="purple")
plt.title("Barplot for Sales column monthly wise")
```

```python
#Barplot for Sub_Category Column
plt.figure(figsize=(22,10))
category=df[["Sub_Category", "Sales"]].groupby(['Sub_Category'], as_index=False).sum().sort_values(by='Sales', ascending=False)
sns.barplot(x='Sub_Category', y='Sales', data=category,color="c")
plt.title("Barplot for Sub_Category Column by Sales")
```

```python
#Correlation Matrix
plt.figure(figsize=(15,6))
numerical=['Sales','Year','Month']
corr_matrix = df[numerical].corr()
sns.heatmap(corr_matrix, annot=True,fmt='.2f',linewidths=.5)
plt.title("Correlation Matrix")
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='Region',y='Sales',hue='Category',errorbar=None)
plt.title("Sales and Region by Category")
```

```python
#Boxplot for Sales VS Category
plt.figure(figsize=(10,8))
sns.boxplot(data=df.head(200),x='Category',y='Sales')
plt.title("Boxplor for Sales VS Category")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='Year',y='Sales',hue='Region',errorbar=None)
plt.title("Sales by Region and Category")
```
