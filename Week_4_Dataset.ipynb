**Importing** **libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading** **the** **dataset** **Population**

```python
path="/content/drive/MyDrive/PrepInsta Week-4/country_population.csv"
population=pd.read_csv(path)
remove_columns=['Indicator Name','Indicator Code']
population=population.drop(columns=remove_columns)
population=population.drop(108)
population.shape
population.columns=population.columns.str.lower().str.replace(" ","_")
population.iloc[:,2:]=population.iloc[:,2:].apply(lambda row: row.fillna(row.median()),axis=1)
population
```

**Loading** **the** **dataset** **Expectancy**

```python
path='/content/drive/MyDrive/PrepInsta Week-4/life_expectancy.csv'
expectancy=pd.read_csv(path)
remove_columns=['Indicator Name','Indicator Code']
expectancy=expectancy.drop(columns=remove_columns)
expectancy=expectancy.drop(108)
expectancy.columns=expectancy.columns.str.lower().str.replace(" ","_")
expectancy.dropna(axis=0,inplace=True)
expectancy
```

**Loading** **the** **dataset** **Fertility**

```python
path='/content/drive/MyDrive/PrepInsta Week-4/fertility_rate.csv'
fertility=pd.read_csv(path)
remove_columns=['Indicator Name','Indicator Code']
fertility=fertility.drop(columns=remove_columns)
fertility=fertility.drop(108)
fertility.columns=fertility.columns.str.lower().str.replace(" ","_")
fertility.dropna(axis=0,inplace=True)
fertility
```

**Loading** **the** **dataset** **Metadata**

```python
path='/content/drive/MyDrive/PrepInsta Week-4/Metadata_Country.csv'
metadata=pd.read_csv(path)
metadata.columns=metadata.columns.str.lower().str.replace(" ","_")
metadata
```

**Merging** **the** **meta** **dataframe** ***to*** **Population** **dataframe**

```python
columns_to_merge=['country_code','region']
population_metadata=population.merge(metadata[columns_to_merge],on='country_code')
population_metadata
```

**Melting** **the** **population** **and** **metadata** **dataframes**

```python
melted_data_population=population_metadata.melt(id_vars=['country_name','country_code','region'],var_name='year',value_name='population')
sorted_data_population=melted_data_population.sort_values(by=['country_name','year'])
sorted_data_population
```

**Melting** **the** **Fertility** **dataframe**

```python
melted_data_fertility=fertility.melt(id_vars=['country_name','country_code'],var_name='year',value_name='fertility')
sorted_data_fertility=melted_data_fertility.sort_values(by=['country_name','year'])
sorted_data_fertility
```

**Merging** **the** **two** **sorted_data_fertility** **dataframe** **to** **sorted_data_population** **dataframe**

```python
columns_to_merge=['fertility','country_name','year']
population_fertility=sorted_data_population.merge(sorted_data_fertility[columns_to_merge],on=['country_name','year'])
population_fertility
```

**Melting** **the** **Expectancy** **dataframe**

```python
melted_data_expectancy=expectancy.melt(id_vars=['country_name','country_code'],var_name='year',value_name='expectancy')
sorted_data_expectancy=melted_data_expectancy.sort_values(by=['country_name','year'])
sorted_data_expectancy
```

**Merging** **the** **two** **sorted_data_expectancy** **dataframe** **to** **sorted_data_population** **dataframe**

```python
columns_to_merge=['expectancy','country_name','year']
population_expectancy=sorted_data_population.merge(sorted_data_expectancy[columns_to_merge],on=['country_name','year'])
population_expectancy
```

**Final Dataframe**

```python
final_df=pd.merge(sorted_data_population,population_fertility,left_on=['country_name','country_code','year','region','population'],right_on=['country_name','country_code','year','region','population'],how='inner').merge(population_expectancy,on=['country_name','country_code','year','region','population'],how='inner')
final_df
df=final_df
df.dropna(subset=['region'],inplace=True)
df.dropna(subset=['population'],inplace=True)
df['fertility']=final_df['fertility'].round(decimals=2)
df['expectancy']=final_df['expectancy'].round(decimals=2)
df['population'].fillna(0,inplace=True)
df['year']=df['year'].astype(int)
df.dropna(axis=0,inplace=True)
df.shape
df
```

**Animated scatter plot created with Plotly Express**

```python
import plotly.express as px
fig=px.scatter(data_frame=df,x='fertility',y='expectancy',animation_frame="year", animation_group="country_name",
           size='population', color="region", hover_name="country_name",template='plotly_dark',
         size_max=65, range_x=[0,10], range_y=[10,100],title='Fertility Rate VS Expectancy Rate')
fig.show()


```

**Bar graph for Region VS Population**
