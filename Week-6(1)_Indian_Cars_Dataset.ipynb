**Cleaning the data**

**Make column**

```python
df['Make'].isnull().value_counts()
df=df.dropna(subset=['Make'])
df['Make'].isnull().value_counts()
```

**Ex_showroom_price column**

```python
df.rename(columns = {'Ex-Showroom_Price':'Ex_Showroom_Price'}, inplace = True)
df['Ex_Showroom_Price']=df['Ex_Showroom_Price'].str.replace("Rs.","")
df['Ex_Showroom_Price']=df['Ex_Showroom_Price'].str.replace(",","")
df['Ex_Showroom_Price']=df['Ex_Showroom_Price'].astype(int)
df['Ex_Showroom_Price'].dtype
```

**Child_safety_Locks Column**

```python
df['Child_Safety_Locks']=df['Child_Safety_Locks'].replace(np.nan,'No')
```

**Fuel_Tank_Capacity column**

```python
df['Fuel_Tank_Capacity']=df['Fuel_Tank_Capacity'].str.replace('litres','')
df['Fuel_Tank_Capacity']=df['Fuel_Tank_Capacity'].replace(np.nan,df.Fuel_Tank_Capacity.median())
df['Fuel_Tank_Capacity']=df['Fuel_Tank_Capacity'].astype(float)

```

**Height, Length and Width columns**

```python
df['Height']=df.Height.str.replace("mm","")
df['Length']=df.Length.str.replace("mm","")
df['Width']=df.Width.str.replace("mm","")
```

**City_Mileage column**

```python
df['City_Mileage']=df['City_Mileage'].str.strip('km/litre')
df['City_Mileage']=df['City_Mileage'].str.replace(",",".")
df['City_Mileage']=df['City_Mileage'].str.replace("?","")
df['City_Mileage']=df['City_Mileage'].str.replace(",",".")
df['City_Mileage']=df['City_Mileage'].str.replace("12.5-12.7","12.5")
df['City_Mileage']=df['City_Mileage'].astype(float)
df['City_Mileage']=df['City_Mileage'].replace(np.nan,df.City_Mileage.median())
```

**Displacement Column**

```python
df['Displacement']=df['Displacement'].str.replace("cc","")
df['Displacement']
```
